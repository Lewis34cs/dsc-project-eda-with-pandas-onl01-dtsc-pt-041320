```python
# Let's get started importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Loading the data
df = pd.read_csv('C:\\Users\\Lewis\\flatiron_coursework\\section04\\house-prices-advanced-regression-techniques\\train.csv')
```

```python
# Investigate the Data
df.info()
```

```python
# Investigating Distributions using scatter_matrix
pd.plotting.scatter_matrix(df[['SalePrice', 'YrSold', 'LotArea']], figsize=(10,10))
```

```python
# Create a plot that shows the SalesPrice Distribution
fig, ax = plt.subplots(figsize=(10,10))
ax.hist(df['SalePrice'], bins=15)
ax.set_xlabel('Price of House')
ax.set_ylabel('Number of Houses')
ax.set_title('Sales Prices of Houses')
df['SalePrice'].describe()
```

```python
# Create a plot that shows the LotArea Distribution
fig, ax = plt.subplots(figsize=(10,10))
ax.hist(df['LotArea'], bins=100)
ax.set_xlabel('Amount of Land per Lot')
ax.set_ylabel('Number of Houses')
ax.set_title('Lot Area Distribution of Houses in Ames')
df['LotArea'].describe()
```

```python
# Create a plot that shows the Distribution of the overall house condition
fig, ax = plt.subplots(figsize=(10,10))
ax.hist(df['OverallCond'], bins=10)
ax.set_xlabel('Condition Level')
ax.set_ylabel('Number of Houses')
ax.set_title('Distribution of Overall House Conditions in Ames')
df['OverallCond'].describe()
```

```python
# Create a Box Plot for SalePrice
fig, ax = plt.subplots(figsize=(10,10))
ax.boxplot(df['SalePrice'])
ax.set_xlabel('')
ax.set_ylabel('Price of House')
ax.set_title('Sale Prices of Houses in Ames')
df['SalePrice'].describe()
```
